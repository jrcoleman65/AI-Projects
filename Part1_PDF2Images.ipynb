```python
# @title FUNCTION: PDF2Images

def PDF2Images(pdf_file: str, output_folder: str):

  if not os.path.exists(output_folder):
    os.makedirs(output_folder, exist_ok=True)

  # convert PDF to images
  images = convert_from_path(pdf_file)

  image_paths = []

  # Save images and paths
  for i, image in enumerate(images):
    image_path = os.path.join(output_folder, f"page{i+1}.jpg")
    image.save(image_path, "JPEG")
    image_paths.append(image_path)

  return image_paths
```

```python
# @title Define Paths & call function

pdf_file =  "/content/drive/MyDrive/Templates/PDF2Images/pdf_file/things-mother-used-to-make.pdf"
output_folder = "/content/drive/MyDrive/Templates/PDF2Images/images/"

image_paths = PDF2Images(pdf_file, output_folder)
```
